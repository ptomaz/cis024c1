** Exercise 2 - Line chart **

Consider the following dataset of the weights of 7 cars and their corresponding fuel consumption

```
weight = [3170,3455,3222,3983,2441,2500,2390]
fuel_consumption = [26,30,28,23,36,33,38]
```

1. Create a line chart that plots the weight of cars against their fuel consumption.
2. Comment on your findings. How does fuel consumption change as the weight of the car increases?

Make sure to add labels and a title to your chart

````python
### YOUR CODE GOES BELOW
import matplotlib.pyplot as plt

weight = [3170,3455,3222,3983,2441,2500,2390]

fuel_mileage = [26,30,28,23,36,33,38]

weight, fuel_mileage = zip(*sorted(zip(weight, fuel_mileage)))

#print weight
#print fuel_mileage
plt.figure(figsize=(12,6))

plt.xlabel("Vehicle Weight\n\nThe general trend is for fuel mileage to down as the\nvehicle weight goes up.  The data is not monotonically decreasing\nthough since car weight is not the only fuel economy factor.")
plt.ylabel("Fuel Mileage")
plt.title(" The Weight of a Vehicle vs. its Fuel Mileage")

plt.plot(weight,fuel_mileage)
#plt.legend()
plt.show()

### END CODE
````
